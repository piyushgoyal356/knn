```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('mnist_train.csv')
# print(data.shape)
data = data.values
```

```python
def imageshow(img, label):
#     img = data[i][1:]
#     label = data[i][0]   
    img = img.reshape((28, 28))
    plt.imshow(img, cmap='Greys')
    plt.show()
    print(label)
```

```python
split = round(0.8*data.shape[0])

X = data[:,1:]
Y = data[:,0].reshape(-1)

X_train = X[:split,:]
Y_train = Y[:split]
X_test = X[split:,:]
Y_test = Y[split:]

# print(X_test.shape)
```

```python
# imageshow(X_train[11],Y_train[11])
def dist(p1,p2):
    return np.sqrt(sum((p1-p2)**2))

def knn(X,Y,testPoint,k=10):
    n = X.shape[0]
    m = X.shape[1]
    dist_matrix = []
    
    for i in range(0,n):
        d = dist(X[i],testPoint)
        dist_matrix.append((d,Y[i]))
        
    dist_matrix = sorted(dist_matrix)
    dist_matrix = np.array(dist_matrix[:k])
    vals,counts = np.unique(dist_matrix,return_counts=True)
    pred = int(vals[np.argmax(counts)])
    
    return pred
```

```python
def accuracy(X,Y):
    n = X.shape[0]
    count = 0
    with open('output.txt','w') as file:
        for i in range(0,n):
    #         a = imageshow(X_test[13],Y_test[13])
            pred = knn(X_train,Y_train,X[i])

            if Y[i]==pred:
                count+=1
                file.write(str(i)+"->pass\n")
                print(str(i)+"->pass")
            else:
                file.write(str(i)+"->fail\n")
                print(str(i)+"->fail")

    acc = count/n*100
    return acc
```

```python
acc = accuracy(X_test[:1000,:],Y_test[:1000])
print("accuracy = " + str(acc) + "%")

# imageshow(X_test[105],Y_test[105])
# knn(X_train,Y_train,X_test[105])
```
